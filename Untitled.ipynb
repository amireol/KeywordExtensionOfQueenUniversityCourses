```python
df['comb'] = df['Course title'] + " " + df['Description']
df['comb'][:3]
```

```python
# Lemmatizating
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize
```

```python
import nltk
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('stopwords')
```

```python
lem = WordNetLemmatizer()
li = []
for sentence in df['comb']:
    word_list = word_tokenize(sentence)
    lemmatized_output = ' '.join([lem.lemmatize(word) for word in word_list])
    li.append(lemmatized_output)
df['comb'] = li
```

```python
# Stemming
from nltk.stem import PorterStemmer
```

```python
ps = PorterStemmer()
df['comb'] = [ps.stem(sentence) for sentence in df['comb']]
```

```python
df['comb'].head()
```

```python
# StopWords
from nltk.corpus import stopwords
```

```python
stop_words = set(stopwords.words('english'))
li = []
for sentence in df['comb']:
    word_tokens = word_tokenize(sentence)
    filtered_sentence = " ".join([w for w in word_tokens if not w in stop_words])
    li.append(filtered_sentence)
  
df['comb'] = li
```

```python
import re
li = []
for sentence in df['comb']:
    regex = re.compile('[^a-zA-Z]')
    reged = regex.sub(' ', sentence)
    word_tokens = word_tokenize(reged)
    woli = []
    for word in word_tokens:
        if len(word) > 2:
            woli.append(word)
    li.append(" ".join(woli))
df['comb'] = li  
```

```python
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import association_rules, apriori
```

```python
items = [word_tokenize(x) for x in df['comb']]
```

```python
te = TransactionEncoder()
te_ary = te.fit(items).transform(items)
```

```python
te_ary.shape
```

```python
dfx = pd.DataFrame(te_ary, columns=te.columns_)
```

```python
dfx.head()
```

```python
frequent_itemsets = apriori(dfx, min_support=0.2, use_colnames=True)
```

```python
association_rules(frequent_itemsets, metric="confidence", min_threshold=0.5)
```
